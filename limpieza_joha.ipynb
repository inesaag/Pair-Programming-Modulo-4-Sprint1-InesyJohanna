```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
guion = pd.read_csv("./files/simpsons_script_lines.csv")
guion.head(3)
```

```python
guion.columns
```

```python
# Columnas a eliminar : 'normalized_text','raw_text','timestamp_in_ms','speaking_line'

def eliminar_columnas(df,column):
    return df.drop(columns=column)
lista_col = ['index','normalized_text','raw_text','timestamp_in_ms','speaking_line']
guion = eliminar_columnas(guion,lista_col)
```

```python
guion.head()
```

```python
episodes= pd.read_csv("./files/simpsons_episodes.csv")
episodes.head(3)
```

```python
# Columnas a eliminar : 'image_url','number_in_season',	'number_in_series',	'original_air_date', 'original_air_year','production_code',	'season', 'video_url', 'views'

lista_col = ['index','image_url', 'number_in_season',	'number_in_series',	'original_air_date', 'original_air_year','production_code',	'season', 'video_url', 'views']
episodes = episodes.drop(lista_col, axis=1)
```

```python
episodes.head()
```

```python
episodes[episodes["id"] == 32]
```

```python
guion[guion["episode_id"] == 32]
```

```python
episodes = episodes.rename(columns={'id': 'episode_id'})
```

```python
episodes.head()
```

```python
guion[guion["episode_id"] == 10]
```

```python
guion_episodes = pd.merge(guion, episodes, on="episode_id",how='outer')
guion_episodes.head()
```

```python
guion_episodes.dtypes
```

```python
guion_episodes.to_csv('files/guion_episodes.csv')
```
